## Ingestão camada Bronze

Carga dos dados brutos do arquivo CSV armazenado na landing zone para a tabela Bronze.

```python
%sql

USE CATALOG dev;
USE SCHEMA brz;

COPY INTO tb_beneficiarios_raw
FROM (
    SELECT
        *,
        current_timestamp() AS dt_ingestao,
        regexp_extract(_metadata.file_path, '[^/]+$', 0) AS nm_arquivo
    FROM '/Volumes/dev/brz/vl_landing/pda-024-icb-TO-2025_08.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header' = 'true',
    'delimiter' = ';'
);
```
