## Transformação camada Gold

Agregação dos dados da camada Silver para geração de métricas analíticas.

```python
%sql

USE CATALOG dev;
USE SCHEMA gld;

TRUNCATE TABLE tb_beneficiarios_agg;

INSERT INTO tb_beneficiarios_agg
SELECT
    nm_razao_social,
    nm_municipio,
    de_faixa_etaria,
    SUM(qt_beneficiario_ativo) AS total_beneficiarios
FROM dev.slv.tb_beneficiarios
GROUP BY
    nm_razao_social,
    nm_municipio,
    de_faixa_etaria;
```
