## Transformação camada Silver

Aplicação de tipagem, limpeza e padronização dos dados provenientes da camada Bronze.

```python
%sql
USE CATALOG dev;
USE SCHEMA slv;

INSERT INTO tb_beneficiarios
SELECT
    id_cmpt_movel,
    CAST(cd_operadora AS INT) AS cd_operadora,
    TRIM(nm_razao_social) AS nm_razao_social,
    nr_cnpj,
    modalidade_operadora,
    sg_uf,
    CAST(cd_municipio AS INT) AS cd_municipio,
    TRIM(nm_municipio) AS nm_municipio,
    tp_sexo,
    de_faixa_etaria,
    de_faixa_etaria_reaj,
    cd_plano,
    tp_vigencia_plano,
    de_contratacao_plano,
    de_segmentacao_plano,
    de_abrg_geografica_plano,
    cobertura_assist_plan,
    tipo_vinculo,
    CAST(qt_beneficiario_ativo AS INT) AS qt_beneficiario_ativo,
    CAST(qt_beneficiario_aderido AS INT) AS qt_beneficiario_aderido,
    CAST(qt_beneficiario_cancelado AS INT) AS qt_beneficiario_cancelado,
    TO_DATE(dt_carga, 'yyyy-MM-dd') AS dt_carga,
    dt_ingestao,
    nm_arquivo
FROM dev.brz.tb_beneficiarios_raw
```
