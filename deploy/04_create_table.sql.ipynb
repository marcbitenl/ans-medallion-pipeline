## Criação das tabelas - Camada Bronze

Definição da tabela de beneficiários na camada Bronze.
Os dados são armazenados no formato bruto, preservando a estrutura original da fonte,
sem aplicação de regras de transformação ou tipagem.

```python
%sql
USE CATALOG dev;
USE SCHEMA brz;
```

```python
%sql
CREATE TABLE IF NOT EXISTS tb_beneficiarios_raw (
    id_cmpt_movel STRING COMMENT 'Identificador do período de competência',
    cd_operadora STRING COMMENT 'Código da operadora',
    nm_razao_social STRING COMMENT 'Nome da operadora',
    nr_cnpj STRING COMMENT 'CNPJ da operadora',
    modalidade_operadora STRING COMMENT 'Modalidade da operadora',
    sg_uf STRING COMMENT 'UF',
    cd_municipio STRING COMMENT 'Código do município',
    nm_municipio STRING COMMENT 'Nome do município',
    tp_sexo STRING COMMENT 'Sexo do beneficiário',
    de_faixa_etaria STRING COMMENT 'Faixa etária',
    de_faixa_etaria_reaj STRING COMMENT 'Faixa etária para reajuste',
    cd_plano STRING COMMENT 'Código do plano',
    tp_vigencia_plano STRING COMMENT 'Tipo de vigência do plano',
    de_contratacao_plano STRING COMMENT 'Tipo de contratação do plano',
    de_segmentacao_plano STRING COMMENT 'Segmentação do plano',
    de_abrg_geografica_plano STRING COMMENT 'Abrangência geográfica do plano',
    cobertura_assist_plan STRING COMMENT 'Cobertura assistencial',
    tipo_vinculo STRING COMMENT 'Tipo de vínculo',
    qt_beneficiario_ativo STRING COMMENT 'Quantidade de beneficiários ativos',
    qt_beneficiario_aderido STRING COMMENT 'Quantidade de beneficiários aderidos',
    qt_beneficiario_cancelado STRING COMMENT 'Quantidade de beneficiários cancelados',
    dt_carga STRING COMMENT 'Data de carga da fonte',
    dt_ingestao TIMESTAMP COMMENT 'Data e hora da ingestão',
    nm_arquivo STRING COMMENT 'Nome do arquivo de origem'
)
USING DELTA
COMMENT 'Tabela Bronze com dados brutos de beneficiários';
```

```python
%sql
describe table dev.brz.tb_beneficiarios_raw
```

## Criação da tabela - Camada Silver

Tabela com dados tratados e tipados, aplicando padronização e regras básicas de qualidade.

```python
%sql
USE CATALOG dev;
USE SCHEMA slv;

CREATE TABLE IF NOT EXISTS tb_beneficiarios (
    id_cmpt_movel STRING COMMENT 'Identificador do período de competência',
    cd_operadora INT COMMENT 'Código da operadora',
    nm_razao_social STRING COMMENT 'Nome da operadora padronizado',
    nr_cnpj STRING COMMENT 'CNPJ da operadora',
    modalidade_operadora STRING COMMENT 'Modalidade da operadora',
    sg_uf STRING COMMENT 'UF',
    cd_municipio INT COMMENT 'Código do município',
    nm_municipio STRING COMMENT 'Nome do município padronizado',
    tp_sexo STRING COMMENT 'Sexo do beneficiário',
    de_faixa_etaria STRING COMMENT 'Faixa etária',
    de_faixa_etaria_reaj STRING COMMENT 'Faixa etária para reajuste',
    cd_plano STRING COMMENT 'Código do plano',
    tp_vigencia_plano STRING COMMENT 'Tipo de vigência do plano',
    de_contratacao_plano STRING COMMENT 'Tipo de contratação do plano',
    de_segmentacao_plano STRING COMMENT 'Segmentação do plano',
    de_abrg_geografica_plano STRING COMMENT 'Abrangência geográfica do plano',
    cobertura_assist_plan STRING COMMENT 'Cobertura assistencial',
    tipo_vinculo STRING COMMENT 'Tipo de vínculo',
    qt_beneficiario_ativo INT COMMENT 'Quantidade de beneficiários ativos',
    qt_beneficiario_aderido INT COMMENT 'Quantidade de beneficiários aderidos',
    qt_beneficiario_cancelado INT COMMENT 'Quantidade de beneficiários cancelados',
    dt_carga DATE COMMENT 'Data de carga da fonte convertida',
    dt_ingestao TIMESTAMP COMMENT 'Data e hora da ingestão',
    nm_arquivo STRING COMMENT 'Nome do arquivo de origem'
)
USING DELTA
COMMENT 'Tabela Silver com dados tratados, tipados e padronizados a partir da camada Bronze';
```

```python
%sql
describe table dev.slv.tb_beneficiarios
```

## Criação da tabela - Camada Gold

Tabela com dados agregados para consumo analítico, organizados para responder perguntas de negócio.

```python
%sql
USE SCHEMA gld;

CREATE TABLE IF NOT EXISTS tb_beneficiarios_agg (
    nm_razao_social STRING COMMENT 'Nome da operadora',
    nm_municipio STRING COMMENT 'Município',
    de_faixa_etaria STRING COMMENT 'Faixa etária',
    total_beneficiarios INT COMMENT 'Total de beneficiários'
)
USING DELTA
COMMENT 'Tabela Gold com agregações para análise de beneficiários';
```
