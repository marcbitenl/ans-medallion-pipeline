## Top 5 operadoras com maior número de beneficiários

Consulta responsável por identificar as operadoras com maior volume de beneficiários,
considerando os dados agregados da camada Gold.

```python
%sql
SELECT
    nm_razao_social,
    SUM(total_beneficiarios) AS total_beneficiarios
FROM dev.gld.tb_beneficiarios_agg
GROUP BY nm_razao_social
ORDER BY total_beneficiarios DESC
LIMIT 5;
```
