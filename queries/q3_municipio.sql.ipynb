## Distribuição de beneficiários por município

Consulta que apresenta a distribuição de beneficiários por município,
permitindo análise geográfica dos dados.

```python
%sql
SELECT
    nm_municipio,
    SUM(total_beneficiarios) AS total_beneficiarios
FROM dev.gld.tb_beneficiarios_agg
GROUP BY nm_municipio
ORDER BY total_beneficiarios DESC;
```
