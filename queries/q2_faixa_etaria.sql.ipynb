## Faixa etária com maior número de beneficiários

Consulta que identifica a faixa etária com maior concentração de beneficiários,
com base na camada Gold.

```python
%sql
-- Faixa etária com maior número de beneficiários
SELECT
    de_faixa_etaria,
    SUM(total_beneficiarios) AS total_beneficiarios
FROM dev.gld.tb_beneficiarios_agg
GROUP BY de_faixa_etaria
ORDER BY total_beneficiarios DESC
LIMIT 1;
```
